## Importando a biblioteca e o dataset

* A biblioteca pandas será utilizada para a manipuação dos datasets
* e o dataset que estaremos utilizando é o "dataset", que contém a relação entre sintomas e as doenças

```python
import pandas as pd


#Symptom-severity.csv
#symptom_Description.csv
#symptom_precaution.csv

dataset = pd.read_csv("dataset.csv")
    
dataset.head(10)
```

## Tratamentos de valores faltantes

* Esse dataset contém muitos valores NaN graças a forma como seus dados estão distribuídos pelas colunas, e muitos desses valores podem atrapalhar os cálculos de machine learning então é melhor se transformarmos cada sintoma disponível em uma coluna e suas linhas representariam se aquela doença tem ou não aquele sintoma, uma operação conhecida como "variáveis dummy"

```python
# Preenchendo todos os NaN com uma string vazia

dataset.fillna("", inplace = True)
```

```python
dataset.shape
```

```python
dataset.head(4)
```

* Abaixo vamos capturar todos os sintomas únicos

```python
points = dataset.Symptom_1.unique()
print(points)
```

```python
sintomas = []
i = 0
#Iterando através das colunas:
for column in dataset:
    if i > 0:
        points = dataset[column].unique()
        for sintoma in points:
            #Adicionando os sintomas encontrados ao mesmo tempo que checa se eles já foram adicionados
            if sintoma not in sintomas:
                sintomas.append(sintoma)
    #Pulando a primeira coluna que apenas diz as doenças e estamos buscando os sintomas
    else:
        i+= 1


sintomas.remove('')

registrosSintomas = []
for i in range(1,4921,1):
    registrosSintomas.append([0] * len(sintomas))


    
```

```python
sintomas
```

```python
for index, row in dataset.iterrows():
    i = 0
    for sintoma in sintomas:
        for column in row:
            if sintoma == column:
                registrosSintomas[index][i] = 1
        i += 1
print(registrosSintomas)
```

* Observe a diferença entre o Dataset original e o modificado com as variáveis dummy

```python
dfSymtoms = pd.DataFrame(data=registrosSintomas, columns=sintomas)
dfSymtoms.insert(0, "Disease",dataset["Disease"] )
dfSymtoms.head(3)
```

## Salvando Json com todos os sintomas 

* No nosso aplicativo precisaremos utilizar os nomes dos sintomas e não uns e zeros para podermos realizar as previsões depois do treinamento 

```python
dicionario = {}
i = 0
for sintoma in sintomas:
    dicionario[i] = sintoma
    i += 1

```

## Treinando a base de dados

```python
#Dividindo base entre classes e previsores
xsim = dfSymtoms.iloc[:,1:132].values
ysim = dfSymtoms.iloc[:,0].values
```

```python
#Dividindo registros entre teste e treinamento
from sklearn.model_selection import train_test_split

x_sim_treinamento, x_sim_teste, y_sim_treinamento, y_sim_teste = train_test_split(xsim,ysim,test_size=0.25,random_state=0)
```

```python
# Salvando as bases separadas
import pickle
with open ("E:\\nlp\\base_symtoms.pkl",mode="wb") as f:
    pickle.dump([x_sim_treinamento, x_sim_teste, y_sim_treinamento, y_sim_teste],f)
```

* Agora finalmente podemos importar a biblioteca de treinamento e treinar a nossa base de dados

```python
#Importando o algortimo de machine Learning Naive Bayes
from sklearn.naive_bayes import GaussianNB
naive_symtoms = GaussianNB()
# Treinando a base de dados
naive_symtoms.fit(x_sim_treinamento,y_sim_treinamento )
```

### Testando a base treinada

```python
entrada = [0] * len(sintomas)
sintomasPaciente = ["itching","skin_rash","nodal_skin_eruptions","dischromic _patches"]

i = 0
for sintoma in sintomas:
    for sinP in sintomasPaciente:
        if sintoma == sinP:
            entrada[i] = 1
    i += 0
    
entrada
```

* Abaixo podemos ver um teste realizado com uma entrada de sintomas convertidos para a configuração de variáveis dummy

```python
previsao = naive_symtoms.predict([entrada])
previsao
```

```python
previsores = naive_symtoms.predict(x_sim_teste)
previsores
```

### Verificando a acurácia do treinamento

```python
#verificando acuracia
from sklearn.metrics import confusion_matrix
confusion_matrix(y_sim_teste,previsores)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_sim_teste,previsores))
```
